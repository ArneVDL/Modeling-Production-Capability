```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import truncnorm

# Set the parameters for the normal distribution
mean = 0
std_dev = 1

# Set the minimum and maximum values for truncation
a, b = -2, 2

# Create a truncated normal distribution
truncated_dist = truncnorm((a - mean) / std_dev, (b - mean) / std_dev, loc=mean, scale=std_dev)

# Generate random samples from the truncated distribution
samples = truncated_dist.rvs(size=1000)

# Plot the histogram of the samples
plt.hist(samples, bins=50, density=True, alpha=0.6, color='g')

# Plot the probability density function (PDF) of the truncated normal distribution
x = np.linspace(-3, 3, 100)
pdf = truncated_dist.pdf(x)
plt.plot(x, pdf, 'r-', lw=2)

plt.title('Truncated Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```
